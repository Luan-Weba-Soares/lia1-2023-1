# Condicional if

```python
# Condicional if (Se)
if 5 > 2:
    print("A sentença é verdadeira!")
```

```python
# Condicional if...else
if 5 < 2:
    print("A sentença é verdadeira!")
else:
    print("A sentença é falsa!")
```

```python
# Condicional if...else com variável
dia = "Quarta-feira"
if dia == 'Sexta-feira':
    print("Hoje fará sol!")
else:
    print("Hoje vai chover!")
```
